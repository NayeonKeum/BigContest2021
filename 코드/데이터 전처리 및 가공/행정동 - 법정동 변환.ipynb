## 결론
- 제주시의 경우, 법정동 -> 행정동 문제 없음
- 서귀포시의 경우, 법정동 내에서 위치에 따라 행정동이 달라지는 문제가 발생함
- ex) 서귀포시 서귀동 같은 경우는 천지동, 송산동, 정방동, 중앙동 4개로 달라질 수 있음

```python
제주시_법정동_행정동 = jeju[jeju["시군구명"] == "제주시"]
서귀포시_법정동_행정동 = jeju[jeju["시군구명"] == "서귀포시"]
```

## 제주시 표제부 법정동 -> 행정동 매핑하기
- 제주시_법정동_행정동 dataframe에서 우리한테 필요한 행정동 코드는 행정기관코드
- 행정동코드.csv : 도두동(50110690)
- 제주시_법정동_행정동.csv : 도두동(5011069000)
- 제주시 시군구 코드 : 50110

```python
제주시표제부 = pd.read_csv("표제부 조회_제주시.csv")[["번호","대지위치","시군구코드","법정동코드","도로명대지위치"]]
제주시표제부 = 제주시표제부.astype(str)
제주시표제부["시군구_법정동코드"] = 제주시표제부["시군구코드"] + 제주시표제부["법정동코드"]
```

```python
print(제주시표제부.shape)
제주시표제부.head(3)
```

```python
제주시_법정동행정동 = 제주시_법정동_행정동[["시군구_법정동코드","행정동명","법정동명","행정기관코드"]]
제주시_법정동행정동.columns = ["시군구_법정동코드","행정동명","법정동명","행정동코드"]
제주시_법정동행정동["행정동코드"] = 제주시_법정동행정동["행정동코드"].apply(lambda x : str(x)[:-2])
제주시_법정동행정동 = 제주시_법정동행정동.astype(str)
```

```python
제주시_법정동행정동.head(3)
```

- 법정동코드로 구성된 "제주시표제부" 데이터
- 법정동코드, 행정동코드로 구성된 "제주시_법정동행정동" 데이터 같이 매핑해주기

```python
제주시표제부_행정동매핑 = pd.merge(left = 제주시표제부, right = 제주시_법정동행정동,
                       on = "시군구_법정동코드", how = "left")
```

```python
제주시표제부_행정동매핑["행정동명"].value_counts().index
```

```python
제주시표제부_행정동매핑
```

제주시 표제부 매핑 끝

```python
제주시표제부_행정동매핑.to_csv("표제부 조회_제주시_행정동매핑.csv",encoding="utf-8-sig",index=False)
```

## 서귀표시 표제부 법정동 -> 행정동
- 중복체크 : 서귀동 / 동홍동 / 서홍동 / 토평동
- 서귀포시의 경우 4개의 법정동이 문제가 된다 나머지는 문제 없음

```python
jeju[jeju.duplicated(["법정동명"],keep=False)].sort_values(by = "법정동명")
```

```python
서귀포_법정동행정동 = jeju[jeju["시군구명"] == "서귀포시"]
서귀포_법정동행정동[서귀포_법정동행정동.duplicated(["법정동명"],keep=False)].sort_values(by = "법정동명")
```

동홍동 / 서귀동 / 서홍동 / 토평동 제외하면 나머지는 행정동 - 법정동 1:1 대응한다
- 서귀포시표제부_4개동제외 : 4개동 제외하고 행정동 매핑해주기

```python
서귀포시표제부 = pd.read_csv("표제부 조회_서귀포시.csv")
서귀포시표제부["법정동"] = 서귀포시표제부["대지위치"].str.split(" ").str[2]
```

```python
서귀포시표제부["시군구_법정동코드"] = 서귀포시표제부["시군구코드"].astype(str) + 서귀포시표제부["법정동코드"].astype(str)
```

```python
cond1 = 서귀포시표제부["법정동"] != "동홍동" ; cond2 = 서귀포시표제부["법정동"] != "서귀동" ;
cond3 = 서귀포시표제부["법정동"] != "서홍동" ; cond4 = 서귀포시표제부["법정동"] != "토평동" ;


서귀포시표제부_4개동제외 =  서귀포시표제부[ cond1 & cond2 & cond3 & cond4]
```

```python
서귀포시표제부_4개동제외.head(3)
```

표제부 데이터와 매핑하기 전에 법정동-행정동 데이터 형식 맞춰주기

```python
서귀포시_법정동_행정동 = 서귀포시_법정동_행정동[["시군구_법정동코드","행정동명","법정동명","행정기관코드"]]
서귀포시_법정동_행정동.rename(columns={"행정기관코드":"행정동코드"},inplace=True)
# 행정동코드로 이름 변경하기

서귀포시_법정동_행정동["행정동코드"] = 서귀포시_법정동_행정동["행정동코드"].apply(lambda x : str(x)[:-2])
서귀포시_법정동_행정동 = 서귀포시_법정동_행정동.astype(str)
```

표제부 데이터와 매핑해주기

```python
서귀포시표제부_4개동제외_행정동매핑 = pd.merge(left = 서귀포시표제부_4개동제외, right = 서귀포시_법정동_행정동,
                       on = "시군구_법정동코드", how = "left")
```

```python
서귀포시표제부_4개동제외_행정동매핑.to_csv("표제부 조회_서귀포시_4개동 제외 행정동매핑.csv",
                           encoding="utf-8-sig",index=False)
```

서귀포시표제부_4개동제외 : 4개동 제외하고 행정동 매핑해주기 성공

### 서귀포시 제외한 4개 법정동 행정동 매핑해주기

```python
cond1 = 서귀포시표제부["법정동"] == "동홍동" ; cond2 = 서귀포시표제부["법정동"] == "서귀동" ;
cond3 = 서귀포시표제부["법정동"] == "서홍동" ; cond4 = 서귀포시표제부["법정동"] == "토평동" ;


서귀포시표제부_4개동 =  서귀포시표제부[ cond1 | cond2 | cond3 | cond4]
```

```python
서귀포시표제부_4개동.to_csv("서귀포시표제부_4개동.csv",encoding="utf-8-sig",index=False)
```
