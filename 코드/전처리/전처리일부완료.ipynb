```python
import pandas as pd
from datetime import datetime
```

```python
idx=pd.date_range("2018-01-01","2021-08-31")
행정동명=["건입동", "구좌읍", "남원읍", "노형동", "대륜동", "대정읍", "대천동", "도두동", "동홍동", 
      "봉개동", "삼도1동", "삼도2동", "삼양동", "서홍동", "성산읍", "송산동", "아라동", "안덕면",
      "애월읍", "연동", "영천동", "예래동", "오라동", "외도동", "용담1동", "용담2동", "우도면", 
      "이도1동", "이도2동", "이호동", "일도1동", "일도2동", "정방동", "조천읍", "중문동", "중앙동", 
      "천지동", "추자면", "표선면", "한경면", "한림읍", "화북동", "효돈동"]

행정시명=["제주시", "제주시", "서귀포시", "제주시", "서귀포시", "서귀포시", "서귀포시", "제주시", "서귀포시", 
      "제주시", "제주시", "제주시", "제주시", "서귀포시", "서귀포시", "서귀포시", "제주시", "서귀포시",
      "제주시", "제주시", "서귀포시", "서귀포시", "제주시", "제주시", "제주시", "제주시", "제주시", 
      "제주시", "제주시", "제주시", "제주시", "제주시", "서귀포시", "제주시", "서귀포시", "서귀포시", 
      "서귀포시", "제주시", "서귀포시", "제주시", "제주시", "제주시", "서귀포시"]
df_idx=[]
for i in range(len(idx)):
    for j in range(len(행정동명)):
        df_idx.append(idx[i])
```

```python
df=pd.DataFrame(range(len(df_idx)))
df['날짜']=df_idx
df['연']=0
df['월']=0
df['행정시']=0
df['행정동']=0
df.drop(0, axis=1, inplace=True)
df
```

```python
for i in range(len(df_idx)):
    df.iloc[i, 1]=df_idx[i].year
    df.iloc[i, 2]=df_idx[i].month
    df.iloc[i, 3]=행정시명[i%len(행정시명)]
    df.iloc[i, 4]=행정동명[i%len(행정동명)]
df
```

```python
행정동별_사업자수_df=pd.read_csv("전처리/행정동별_사업자수(2019).csv")
행정동별_사업자수_df
```

```python
df['사업체수']=0
df
```

```python
for i in range(len(df)):
    for j in range(len(행정동별_사업자수_df)):
        if df.loc[i, '행정동']==행정동별_사업자수_df.loc[j,'행정구역명']:
            df.loc[i, '사업체수']= 행정동별_사업자수_df.loc[j,'사업체수']
df
```

```python
월별_행정동별_세대수_df=pd.read_csv("전처리/월별_행정동별_세대수.csv", encoding="cp949")
월별_행정동별_세대수_df
```

```python
df['세대수']=0
df['세대당 인구']=0
df['남여 비율']=0
df
```

```python
for i in range(len(df)):
    for j in range(len(월별_행정동별_세대수_df)):
        if df.loc[i, '월']==월별_행정동별_세대수_df.loc[j,'기준월'] and df.loc[i, '연']==월별_행정동별_세대수_df.loc[j,'기준년'] and df.loc[i, '행정동']==월별_행정동별_세대수_df.loc[j, '행정구역명']:
            df.loc[i, '세대수']= 월별_행정동별_세대수_df.loc[j,'세대수']
            df.loc[i, '세대당 인구']= 월별_행정동별_세대수_df.loc[j,'세대당 인구']
            df.loc[i, '남여 비율']= 월별_행정동별_세대수_df.loc[j,'남여 비율']
df
```

```python
행정동별_주택종류_df=pd.read_csv("전처리/행정동별_주택종류.csv", encoding="cp949")
행정동별_주택종류_df
```

```python
df['단독주택']=0
df['창고시설']=0
df['제2종근린생활시설']=0
df['제1종근린생활시설']=0
df['공동주택']=0
df['동.식물관련시설']=0
df['숙박시설']=0
df['교육연구시설']=0
df['관광휴게시설']=0
df['분뇨.쓰레기처리시설']=0
df
```

```python
for i in range(len(df)):
    for j in range(len(행정동별_주택종류_df)):
        if df.loc[i, '행정동']==행정동별_주택종류_df.loc[j, '행정구역명']:
            df.loc[i, '단독주택']= 행정동별_주택종류_df.loc[j,'단독주택']
            df.loc[i, '공동주택']= 행정동별_주택종류_df.loc[j,'공동주택']
            df.loc[i, '창고시설']= 행정동별_주택종류_df.loc[j,'창고시설']
            df.loc[i, '제2종근린생활시설']= 행정동별_주택종류_df.loc[j,'제2종근린생활시설']
            df.loc[i, '제1종근린생활시설']= 행정동별_주택종류_df.loc[j,'제1종근린생활시설']
            df.loc[i, '동.식물관련시설']= 행정동별_주택종류_df.loc[j,'동.식물관련시설']
            df.loc[i, '숙박시설']= 행정동별_주택종류_df.loc[j,'숙박시설']
            df.loc[i, '교육연구시설']= 행정동별_주택종류_df.loc[j,'교육연구시설']
            df.loc[i, '관광휴게시설']= 행정동별_주택종류_df.loc[j,'관광휴게시설']
            df.loc[i, '분뇨.쓰레기처리시설']= 행정동별_주택종류_df.loc[j,'분뇨.쓰레기처리시설']
df
```
